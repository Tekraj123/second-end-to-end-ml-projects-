```python
categorical_cols = df.select_dtypes(include=['object']).columns
print(categorical_cols)
```

```python
numerical_cols = df.select_dtypes(exclude=['object']).columns
print(numerical_cols)   
```

```python
one_hot_encoded = OneHotEncoder(handle_unknown='ignore')
scaler = StandardScaler()
# imputer = SimpleImputer(strategy='most_frequent')
```

```python
cat_pipeline = Pipeline(
      steps=[
    #     ('imputer', imputer),
        ('one_hot_encoder', one_hot_encoded)
    ]
)
```

```python
num_pipeline = Pipeline(
    
    steps=[
        # ('imputer', imputer),
        ('scaler',scaler)
    ]
)
```

```python
processor = ColumnTransformer(
    [
    ('num_pipeline', num_pipeline, numerical_cols),
    ('cat_pipeline', cat_pipeline, categorical_cols)
    ]
)
```

```python
processed_Train_data = processor.fit_transform(X_Train)
print(processed_Train_data)
```

```python
processed_Test_data = processor.transform(X_Test)  
```

```python
logistic_model = RandomForestClassifier()
logistic_model.fit(processed_Train_data, Y_Train)  
logistic_predictions = logistic_model.predict(processed_Test_data)
print(classification_report(Y_Test, logistic_predictions))
```
